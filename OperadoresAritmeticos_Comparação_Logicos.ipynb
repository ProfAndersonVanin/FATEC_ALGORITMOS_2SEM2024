A ordem de precedência dos operadores lógicos, quando as expressões têm operadores de mais de uma categoria, é a seguinte: Operadores aritméticos, Operadores de comparação, Operadores lógicos

**1º ==>** `( )` Parênteses

**2º ==>** Exponenciação e radiciação

**3º ==>** Multiplicação, divisão (o que aparecer primeiro)

**4º ==>** `+` ou `–` (o que aparecer primeiro)
